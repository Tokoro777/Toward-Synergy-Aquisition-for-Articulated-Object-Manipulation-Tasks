```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib import cm
object_list = ["apple", "banana", "beerbottle", "box", "random", "log_num_axis3", "log_num_axis5"]
target_file = "/home/hwipsynergy/policy/{}/grasp_dataset_{}.npy"
x_tick = list(range(0, 100, 5))
```

```python
for object in object_list:
    contribution = []
    pos_nums = []
    for i in range(0, 100, 5):
        postures = np.load(target_file.format(object, str(i)))
        if len(postures) < 21:
            contribution.append([0, 0, 0, 0, 0])
            pos_nums.append(0)
        else:
            pos_nums.append(len(postures))
            pca = PCA()
            pca.fit(postures)
            contribution.append(pca.explained_variance_ratio_)

    fig, ax1 = plt.subplots()  # Create a figure containing a single axes.
    ax1.set_title(object)
    ax2 = ax1.twinx()
    ax1.plot(x_tick, [sum(a[:5]) for a in contribution], color=cm.Set1.colors[1], label="top 5 contribution ratio")  # Plot some data on the axes.
    ax1.set_ylim(0, 1.0)
    ax2.plot(x_tick, pos_nums, color=cm.Set1.colors[0], label="postures")
    ax1.legend()
    ax2.legend()
```
